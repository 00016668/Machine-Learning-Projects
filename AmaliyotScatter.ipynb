```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize = (10, 6))
sns.histplot(data=df, x=df['price'])
plt.show()
```

```python
sns.displot(data=df, x=df['mileage'], hue = 'transmission', height = 5, aspect = 2, multiple = 'stack')
plt.show()
```

```python
sns.displot(data=df, x='price', col='transmission', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, kind='kde', x = 'price', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, kde=True, x = 'price', height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize = (12, 6))
sns.kdeplot(data=df, x='price', hue = 'fuelType', shade=True)
plt.show()
```

```python
sns.displot(data=df, x='price', hue='fuelType', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='price', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='mileage', height=5, aspect=2, kind='ecdf')
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='price', hue='fuelType', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.ecdfplot(data=df, x='price', hue='transmission')
plt.grid()
plt.show()
```

```python
sns.rugplot(data=df, x='price', hue='transmission')
plt.grid()
plt.show()
```
